```python
# Importing the library

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data usign pandas

df = pd.read_csv(path)
```

**Task three**: choose the right normalization technique for each variable and normalize it

```python
# create variables fro the sepal length and width

sepal_length = df['sepal_length']
sepal_width = df['sepal_width']
```

```python
# create variables fro the petal length and width

petal_length = df['petal_length']
petal_width = df['petal_width']
```

```python
# standarize of the sepal lenght and the petal width variables

sepal_length_stand = (sepal_length - np.mean(sepal_length))/np.std(sepal_length)
petal_width_stand = (petal_width - np.mean(petal_width))/np.std(petal_width)
```

```python
# normalization of the sepal width and the petal lenght variables

sepal_width_norm = (sepal_width - min(sepal_width))/(max(sepal_width) - min(sepal_width))
petal_length_norm = (petal_length - min(petal_length))/(max(petal_length) - min(petal_length))
```

**Task four**: make 2D plot

```python
# create a new dataset called df2 with the new variables

df2 = pd.DataFrame(dict(sepal_lenght_std = sepal_length_stand,sepal_width_norm = sepal_width_norm, \
                        petal_lenght_norm = petal_length_norm, petal_width_std = petal_width_stand, type_of = df['class']),
                   columns= ['sepal_lenght_std','sepal_width_norm','petal_lenght_norm','petal_width_std','type_of'])
```

```python
# creating a new variables for each type of iris

iris_setosa_df = df2.loc[df2['type_of']=='Iris-setosa']
iris_versicolor_df = df2.loc[df2['type_of']=='Iris-versicolor']
iris_virginica_df = df2.loc[df2['type_of']=='Iris-virginica']
```

```python
# plotting using scatter sepal_lenght_std - sepal_width_norm

plt.scatter(iris_setosa_df['sepal_lenght_std'], iris_setosa_df['sepal_width_norm'])
plt.scatter(iris_versicolor_df['sepal_lenght_std'], iris_versicolor_df['sepal_width_norm'])
plt.scatter(iris_virginica_df['sepal_lenght_std'], iris_virginica_df['sepal_width_norm'])

plt.xlabel('sepal_lenght_std')
plt.ylabel('sepal_width_norm')
plt.title('sepal_lenght_std - sepal_width_norm')
plt.legend(title = 'type',
           colors = ['red','pink']
           labels=['iris-setosa','iris-versicolor','iris-virginica'],
           loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
# plotting using scatter sepal_lenght_std - petal_lenght_norm

plt.scatter(iris_setosa_df['sepal_lenght_std'], iris_setosa_df['petal_lenght_norm'])
plt.scatter(iris_versicolor_df['sepal_lenght_std'], iris_versicolor_df['petal_lenght_norm'])
plt.scatter(iris_virginica_df['sepal_lenght_std'], iris_virginica_df['petal_lenght_norm'])

plt.xlabel('sepal_lenght_std')
plt.ylabel('petal_lenght_norm')
plt.title('sepal_lenght_std - petal_lenght_norm')
plt.legend(title = 'type',
           labels=['iris-setosa','iris-versicolor','iris-virginica'],
           loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.show()
```
