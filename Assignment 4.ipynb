**Task three**: choose the right normalization technique for each variable and normalize it

```python
sepal_length = df['sepal_length']
sepal_width = df['sepal_width']

petal_length = df['petal_length']
petal_width = df['petal_width']
```

```python
sepal_length_stand = (sepal_length - np.mean(sepal_length))/np.std(sepal_length)
petal_width_stand = (petal_width - np.mean(petal_width))/np.std(petal_width)
```

```python
sepal_width_norm = (sepal_width - min(sepal_width))/(max(sepal_width) - min(sepal_width))
petal_length_norm = (petal_length - min(petal_length))/(max(petal_length) - min(petal_length))
```

**Task four**: make 2D plot

```python
df2 = pd.DataFrame(dict(sepal_lenght_std = sepal_lenghth_stand,sepal_width_norm = sepal_width_norm, \
                        petal_lenght_norm = petal_length_norm, petal_width_std = petal_width_stand, cl = df['class']))
```

```python
df2
```

```python
color_dict = {'Iris-setosa':'grey', 'Iris-versicolor':'pink', 'Iris-virginica':'green' }
```

```python
color_dict
```

```python
plt.scatter(df2['sepal_lenght_std'], df2['sepal_width_norm'], color=[color_dict[i] for i in df2['cl']])

plt.xlabel('sepal_lenght_std')
plt.ylabel('sepal_width_norm')
plt.legend(df2['cl'])
plt.show()
```
